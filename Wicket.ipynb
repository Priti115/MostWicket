## Very Very Important

```python
def calculate_span(Span):
    start_year, end_year = map(int, Span.split('-'))
    return end_year - start_year

df['Span_length'] = df['Span'].apply(calculate_span)
df.loc[df['Span_length'].idxmax(),['Player','Span','Span_length']]
```
